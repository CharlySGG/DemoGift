```python
import numpy as np
from scipy.integrate import solve_ivp, odeint

import matplotlib.pyplot as plt
%matplotlib inline
```

1. Resolver numéricamente con `solve_ivp` la siguiente ecuación diferencial:
$$\frac{dx}{dt}=\frac{x}{t+1}$$

```python
#Definimos la ED a resolver
def ED1(x,t):
    return (x/(t+1))
```

Grafique con condición inicial $x_0 = 2$

```python
#Definimos nuestras condiciones iniciales
tspan = [0, 50] #Rango de tiempo
x0 = np.array([2]) #Condicion inicial

#Resolvemos el sistema
sol_x = solve_ivp(ED1, tspan, x0, max_step=0.1)
```

```python
#Graficamos
plt.figure()
plt.plot(sol_x.t, sol_x.y[0, :])
plt.xlabel("Tiempo")
plt.ylabel("Posicion")
plt.grid()
```
